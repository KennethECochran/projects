Apply K-Nearest Neighbor classifier to the iris dataset. Select any TWO of the four variables (sepal length,
sepal width, petal length, petal width).

```python
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

iris = datasets.load_iris() #Loads the iris dataset
X = iris.data[:, 2:4] #grabs the petal length and petal width columns  
y = iris.target
target_names = iris.target_names

X_train, X_test, Y_train, Y_test = train_test_split(X,y, test_size=0.20) #splits the dataset into testing and training. Uses 20% of data for testing
classifier = KNeighborsClassifier() #creates the kneighbors classifier object
classifier.fit(X_train, Y_train) #fits the classifier to the training data
y_pred = classifier.predict(X_test) #predicts the classification for the X_test data

correct_prediction = [] #empty list to hold the correct predictions
incorrect_prediction = [] #empty list to hold the incorrect predictions

for i in range(len(y_pred)): #for loop to check if the classification prediction matches the true classification
    if (y_pred[i] == Y_test[i]):
        correct_prediction.append(i)
    else:
        incorrect_prediction.append(i)

fig = plt.figure(figsize=(10,15))
ax = fig.add_subplot(2, 1, 1) 
II = (y==0)
ax.scatter(X[II,0], X[II, 1], color = "white", edgecolor='red') #plots the Setosa data
II = (y==1)
ax.scatter(X[II,0], X[II, 1], color = "white", edgecolor='green') #plots the versicolor data
II = (y==2)
ax.scatter(X[II,0], X[II, 1], color = "white", edgecolor='blue') #plots the virginica data 
ax.scatter(X_test[correct_prediction, 0], X_test[correct_prediction, 1], color = 'black', marker = '*') #plots the correct predictions
ax.scatter(X_test[incorrect_prediction, 0], X_test[incorrect_prediction, 1], color = 'red', marker = '*') #plots the incorrect predicions

ax.set_title('KNN prediction for Petal Length vs. Petal Width')
ax.set_xlabel('Petal Length')
ax.set_ylabel('Petal Width')
ax.legend(labels = ("Setosa","Versicolor","Virginica","Correct Prediction","Incorrect prediction"))
fig.savefig('../results/KNN_classifier.png')
fig.show() #shows the final scatter plot
```

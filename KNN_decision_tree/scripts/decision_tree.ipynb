Apply sklearn decision tree classifier to the iris dataset

```python
import os
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

iris = datasets.load_iris() #Loads in the iris data set for analysis
X = iris.data
y = iris.target
prediction_accuracy = [] #empty list to hold the K-nearest-neighbor prediction accuracy

for x in range(10): #for loop to train the knn model and get prediction accuracy ten times. 
    X_train, X_test, Y_train, Y_test = train_test_split(X,y, test_size=0.20) #splits the iris data into testing and training, with 20% of the data used for testing
    clf = tree.DecisionTreeClassifier(criterion = 'gini', max_depth = 4) #creates the decision tree classifier object using the gini criteria and a max depth of 4
    clf.fit(X_train,Y_train) #trains the tree 
    Y_pred = clf.predict(X_test) #predicts the classification of the testing data
    prediction_accuracy.append(accuracy_score(Y_test, Y_pred)) #measures the accuracy of the prediction by comparing it to the true classification and then appends it to the list

print("Prediction accuracy:", prediction_accuracy)
print()
print("Mean prediction accuracy:", np.mean(prediction_accuracy))

fig = plt.figure(figsize=(20,20))
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = iris.target_names, filled = True)
fig.savefig('../results/Decision_tree.png')
fig.show()
```
